```python
#import necessary libraries
import pandas as pd
from datetime import datetime , timedelta
```

```python
#import dataset
zulo_bank = pd.read_csv('https://raw.githubusercontent.com/Ahmed9667/Zulo-Bank-RBDMS-PostgreSql/refs/heads/main/zulo_bank.csv')
zulo_bank.head(5)
```

```python
zulo_bank.dropna(axis=0 , inplace=True)
```

```python
zulo_bank.info()
```

Convert to 1NF

```python
# Split FullName to first_name and last_name
zulo_bank['first name'] = zulo_bank['FullName'].str.split(' ').str.get(0)

zulo_bank['last name'] = zulo_bank['FullName'].str.split(' ').str.get(1)
zulo_bank.head(3)

# another solution
#zulo_bank[['first name','last name']]  = zulo_bank['FullName'].str.split(expand=True)
```

Convert 1NF to 2NF

```python
zulo_bank.columns
```

```python
len(zulo_bank)
```

```python
range(1,len(zulo_bank['first name'])+1)
```

```python
#create table customer

customer= zulo_bank[['first name', 'last name','Email', 'Phone']].copy().drop_duplicates().reset_index(drop=True)

#add customer_id column
customer['customer_id'] = range(1,len(customer)+1)
customer = customer[['customer_id','first name', 'last name','Email', 'Phone']]
print(customer.shape)
customer.head(5)
```

```python
#create table transactions

transactions = zulo_bank[[ 'TransactionType', 'Amount',
'TransactionDate']].copy().drop_duplicates().reset_index(drop=True)

#add transaction_id column
transactions['transaction_id'] = range(1,len(transactions)+1)
transactions=transactions[['transaction_id','TransactionType', 'Amount',
'TransactionDate']]

print(transactions.shape)
transactions.head(5)
```

```python
#create accounts type
accounts = zulo_bank[['AccountType', 'Balance', 'OpeningDate']].copy().drop_duplicates().reset_index(drop=True)

#add account_id column
accounts['account_id']= range(1,len(accounts)+1)

accounts= accounts[['account_id','AccountType', 'Balance', 'OpeningDate']]
print(accounts.shape)
accounts.head(5)
```

```python
#create table loans
loans = zulo_bank[['LoanAmount', 'LoanType', 'StartDate', 'EndDate', 'InterestRate']].copy().drop_duplicates().reset_index(drop=True)

#create column loan_id
loans['loan_id'] = range(1,len(loans)+1)
loans=loans[['loan_id','LoanAmount', 'LoanType', 'StartDate', 'EndDate', 'InterestRate']]
print(loans.shape)
loans.head(5)
```

```python
#create zulo bank table

zulo_bank = zulo_bank.merge(customer,on=['first name', 'last name','Email', 'Phone'],how='left')\
            .merge(transactions, on=['TransactionType', 'Amount','TransactionDate'],how='left')\
            .merge(accounts,on=['AccountType', 'Balance', 'OpeningDate'],how='left')\
            .merge(loans,on=['LoanAmount', 'LoanType', 'StartDate', 'EndDate', 'InterestRate'],how='left')\
            [['customer_id','transaction_id','account_id','loan_id']]
            
            
zulo_bank
                                                                                    
```

From above we can define start_date as 2019-01-01 and current date as 2024-02-29

```python
start_date = datetime(2019,1,1)
current_date = datetime(2060,12,31)
num_of_days = (current_date - start_date).days  #will equal to 12783 days

date_list = [start_date + timedelta(days = x) for x in range(num_of_days + 1) ]

print(len(date_list))
date_list
```

```python
# Creat dataframe (table) date
date = {'date_id' : [x for x in range(1 , len(date_list)+1)] ,
        'date': date_list}
date_dim = pd.DataFrame(date)
date_dim.head(3)
```

```python
date_dim['year'] = date_dim['date'].dt.year
date_dim['month'] = date_dim['date'].dt.month
date_dim['day'] = date_dim['date'].dt.day
date_dim.head(5)        
```

```python
date_dim['date'].info()
```

```python
# convert accounts table to 3NF

#convert accounts['OpeningDate'] from object to datetime
date_dim['date'] = pd.to_datetime(date_dim['date'])
accounts['OpeningDate'] = pd.to_datetime(accounts['OpeningDate'])

accounts = accounts.merge(date_dim ,left_on= 'OpeningDate', right_on='date', how='inner' )\
           .rename(columns = {'date_id' : 'OpeningDate_ID'})\
           .reset_index(drop = True)\
           [['account_id','AccountType', 'Balance','OpeningDate_ID']]

accounts.head(5)
```

```python
# Convert transactions table to 3NF
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

transactions = transactions.merge(date_dim , left_on= 'TransactionDate',right_on = 'date' , how = 'inner')\
               .reset_index(drop=True)\
               [['transaction_id','TransactionType', 'Amount','date_id']]
transactions.head(3)
```

```python
transactions=transactions.rename(columns = {'date_id':'TransactionDate_ID'})
transactions.head(3)
```

```python
# Convert table loans

date_dim['date'] = pd.to_datetime(date_dim['date']).dt.date
loans['StartDate'] = pd.to_datetime(loans['StartDate']).dt.date
loans['EndDate']= pd.to_datetime(loans['EndDate']).dt.date

loans = loans.merge(date_dim , left_on='StartDate',right_on='date',how='inner')\
        .rename(columns={'date_id' : 'StartDate_ID'})\
        .merge(date_dim , left_on='EndDate',right_on='date',how='inner')\
        .rename(columns={'date_id' : 'EndDate_ID'})\
        .reset_index(drop=True)\
        [['loan_id','LoanAmount', 'LoanType', 'StartDate_ID', 'EndDate_ID', 'InterestRate']]

loans
```

### Transaction Warehouse Modeling

```python
# create transaction fact table

transactions_dim = transactions[['transaction_id','TransactionType']].copy().drop_duplicates().reset_index(drop=True)
accounts_dim = accounts[['account_id','AccountType', 'Balance']].copy().drop_duplicates().reset_index(drop=True)

transaction_fact_table = zulo_bank.merge(transactions , on='transaction_id',how='inner')\
                                  .merge(accounts , on='account_id' , how='inner')\
                                  [['transaction_id','account_id','OpeningDate_ID','TransactionDate_ID','Amount']]
transaction_fact_table
```

### Loan Warehouse Modeling

```python
customer_dim = customer[['customer_id','first name', 'last name','Email', 'Phone']].copy().drop_duplicates().reset_index(drop=True)
loan_dim = loans[['loan_id','LoanType']].copy().drop_duplicates().reset_index(drop=True)

loan_fact_table = zulo_bank.merge(loans, on='loan_id' , how='inner')\
                           .merge(customer, on='customer_id', how='inner')\
                           [['loan_id','customer_id','StartDate_ID', 'EndDate_ID','LoanAmount','InterestRate']]

loan_fact_table 
```
